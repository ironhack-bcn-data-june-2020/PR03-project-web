```python
#Aquí no te va a funcionar hacer print(json.dumps(myHR_follow._json, indent = 2)) pq ahora estás mirando no a UN user sino a un GRUPO de followers. 
#Tendrás que hacer un loop:

for user in myHR_follow:
    print(json.dumps(user._json, indent = 2))
    
#Aquí ves los followers de la cuenta de Twitter de myHRfuture
```

##### Para obtener el resto de los usuarios hay otra manera de pedir la info, usando el Cursor

Tweepy tiene una clase cursor puedes obtener los items que quieres sin preocuparte por la paginación:

```python



#A la clase (tweepy.Cursor) le pasas el método -sin llamarlo- (api.followers) 
#Le pasas los parámetros del método al que llamo (screen_name = "myHRfuture")
#Le dices a Cursos cuántos items quieres obtener (.items(el número que quieres))
#Iteras el cursor (for user in)
#Dentro del for recuperas el print de los usuarios del json

for user in tweepy.Cursor(api.followers, screen_name = "myHRfuture").items(50):
    print(json.dumps(user._json, indent = 2))
```

#### Cómo sacar los followees de una cuenta de Twitter:

```python
#Vamos a usar el método cursor otra vez para sacar los followees, o sea a quién sigue myHRfuture
#sólo hay que cambiar de api.followers a api.friends

for user in tweepy.Cursor(api.friends, screen_name = "myHRfuture").items(50):
    print(json.dumps(user._json, indent = 2))
```

#### Cómo obtener el timeline de un usuario (todos los tweets que ha hecho un usuario):

```python
#Vamos a usar el cursor otra vez para no tener límites de paginación
#Cambias el método (api.user_timeline)
#Añades otro parámetro, tweet_mode, que determina la longitud del tweet que quiero obtener. Le pones extended para obtener todo el contenido de los tweets de más de 140 caractéres
#También he cambiado en el for "user" por "tweet" para que se adapte a nuestro request

for tweet in tweepy.Cursor(api.user_timeline, screen_name = "myHRfuture", tweet_mode = "extended").items(10):
    print(json.dumps(tweet._json, indent = 2))
```

```python
#El texto del tweet está en la clave "full_text"
#Te devuelve MUCHA información de cada tweet, como por ejemplo el usuario que le dió like o lo retwiteó
```

#### Cómo buscar tweets por palabra(s) clave:

YYYYASSSSSS lo que estábamos buscandoooooo!!! <3 <3

```python
#Aquí el método es api.search
#Como parámetro, si lo que buscas es texto, usas q = "palabra clave"

#Para guardar el contenido del for en una variable, guardo todas las iteraciones del for en un string 

jt = ""
for tweet in tweepy.Cursor(api.search, q = "People Analytics", tweet_mode = "extended").items(100):
    jt += json.dumps(tweet._json, indent = 2)
```

### Este dataframe no nos es útil, así que he empezado otra línea de investigación

```python
#Este método te devuelve tweets con tu palabra clave y con palabras relacionadas...!
```

```python
#Vamos a coger únicamente el full text de los tweets que nos devuelve:

lista = []
for tweet in tweepy.Cursor(api.search, q = "People Analytics", tweet_mode = "extended").items(100):
    lista.append(tweet._json["full_text"])

lista
```

```python
cols = "Contenido"
df = pd.DataFrame(lista, columns = [cols])
```
