### Diccionario y DataFrame topic name - topic id

#### DataFrame

```python
topics = pd.DataFrame()
tname = []
tid = []
tnr = []
driver = webdriver.Firefox()
driver.get(url)
topics = driver.find_elements_by_class_name('topic')
for i in topics:
    try:
              
        tname.append(re.sub(' \(\d*\)','',i.text)) # topic name
        tid.append(re.findall('topic=(\d*)', i.get_attribute('href'))[0]) # topic id
        tnr.append(re.findall('\((\d*)\)', i.text)[-1]) # quantity of movies

        
    except:
        break
driver.quit()
```

```python
topics = pd.DataFrame(index = tid, data = {'name': tname, 'qty': tnr})
```

```python
topics
```

```python
# En función del número de pelis de cada topic, sacamos el número de páginas que habría que scrapear

# Cada página tiene 50 películas como máximo

for i,c in topics.iterrows():
    if c['qty'] != '':
        topics.loc[i, 'pages'] = int((int(c['qty'])-(int(c['qty'])%50))/50+1)
    else:
        topics.loc[i, 'pages'] = 0

```

```python
topics['pages'] = topics['pages'].astype('int')
```

```python
topics
```

# Extracción de las 50 mejores películas de cada uno de los subgéneros

No hemos llevado a cabo la extracción de las 310.176 películas porque a partir de 20.000 aprox. obtenemos un error 429 "too many requests". El enfoque que le daremos para este proyecto será scrapear las 50 películas más votadas de cada uno de los subgéneros y a continuación eliminar duplicados por si hubiera alguna película en más de un subgénero.

Para ver el enfoque que habría que darle al proyecto para extraer todas las películas, consultar el notebook "Filmaffinity v2" en esta misma carpeta

```python
pelisdb = []
```

```python
urltopic = 'https://www.filmaffinity.com/es/movietopic.php'
```

```python
for i,c in tqdm(topics.iterrows()):
    soup = BeautifulSoup(requests.get(urltopic, params={'topic': f'{i}', 'attr': 'top_rated'}).text, 'html.parser')
    pelis = soup.findAll('div', attrs={'class': 'record'})
    for x in pelis:
        pelisdb.append({'Título': x.find('div', attrs={'class': 'mc-title'}).find('a').text,
           'Año': x.find('div', attrs={'class': 'mc-title'}).text.replace(' ','')[-5:-1],
           'Director/es': x.find('div', attrs={'class': 'mc-director'}).text.replace(' (Creador)','').split(', '),
           'Reparto': x.find('div', attrs={'class': 'mc-cast'}).text.split(', '),
           'Género/s': x.find('div', attrs={'class': 'related-topics'}).text.split('\n')[1:-1],
           'País': x.find('img', attrs={'class': 'nflag'}).get_attribute_list('title')[0],
           'Puntuación': x.find('div', attrs={'class': 'avg-w'}).text.replace(',','.'),
           'Nº votos': x.find('div', attrs={'class': 'votes2'}).text.replace('.','')[:-1]
            })
```

```python
pelisdf = pd.DataFrame(pelisdb)
pelisdf
```

## Drop duplicates

A modo de curiosidad, vemos como mientras extraíamos los datos, ha cambiado el nº de votos de 4 películas. Probablemente las películas aparezcan en más de un 'topic' y hayan recibido alguna votación en el rango de tiempo mientras se extraían en ambos topics.

Si eliminamos los duplicados sólo por título, año y país, obtenemos 1.511 películas:

```python
pelis_exc = pelisdf.drop_duplicates(['Título', 'Año','País', 'Nº votos'])
pelis_exc
```

Sin embargo, si lo hacemos sin tener en cuenta el nº de votos, aparecen 1.507, 4 películas menos:

```python
pelis_exc.duplicated(['Título', 'Año','País']).value_counts()
```

```python
pelis_exc[pelis_exc.duplicated(['Título', 'Año','País'])]
```

No sólo podemos ver la diferencia de 1 voto en 3 de las 4 películas, sino que además vemos que la película restante es completamente diferente, por lo que a la hora de eliminar duplicados, filtraremos también por puntuación ya que, a pesar del voto de diferencia, vemos que ésta no ha cambiado para las otras 3 películas que sí son las mismas.

## Tipo de realización

```python
# Extraemos el tipo principal de realización que es

for i,c in pelis.iterrows():
    try:
        tipo = re.findall('\(([^)]*)\)', c['Título'])[-1]
        if tipo == 'C':
            pelis.loc[i, 'Tipo'] = 'Cortometraje'
        elif tipo == 'Serie de TV':
            pelis.loc[i, 'Tipo'] = tipo
        elif tipo == 'TV':
            pelis.loc[i, 'Tipo'] = tipo
        elif tipo == 'Miniserie de TV':
            pelis.loc[i, 'Tipo'] = tipo
        elif tipo == 'Vídeo musical':
            pelis.loc[i, 'Tipo'] = tipo
        elif tipo == 'Ep':
            pelis.loc[i, 'Tipo'] = 'TV'
            
    except:
        pass
```

```python
pelis['Tipo'].fillna('Película', inplace=True)
```
