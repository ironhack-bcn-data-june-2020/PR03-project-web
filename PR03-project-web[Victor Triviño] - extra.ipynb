```python
#Importing all libraries necessary for the project

import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
import urllib.request
import datetime
import warnings
warnings.filterwarnings('ignore')
from bs4 import BeautifulSoup
```

# API

```python
#Setting the url from the API selected

URL = f'https://public.opendatasoft.com/api/records/1.0/search/?dataset=atp&q=&lang=en&rows=10000&facet=tournament&facet=date&facet=series&facet=court&facet=surface&facet=best_of&facet=winner&facet=round&refine.round=The+Final&refine.series=Grand+Slam'

headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.105 Safari/537.36'}

req = requests.get(URL, headers=headers)
```

```python
#Checking the status code return is correct

req.status_code
```

```python
#Convert the req into a json and so be able to analyse it better

results = req.json()

results
```

```python
#We normalize the results into a dataframe as a previous step for getting the dataframe for cleaning

results_flatten = pd.json_normalize(results)

results_flatten
```

```python
#We normalize again the results by records. This will be the dataframe used for cleaning

results_flatten_records = pd.json_normalize(results_flatten.records[0])

results_flatten_records
```

```python
#We modify the dataframe by only selecting the columns we are interested in

results_total = results_flatten_records[['fields.winner', 'fields.date', 'fields.location']]
```

```python
#We filter the data by field location Melbourne (the equivalent to Australian Open)

API_AustralianOpen = results_total[results_total['fields.location'] =='Melbourne']

API_AustralianOpen.sort_values(by='fields.date', ascending=False)
```

```python
#We drop the column 'fields location' as we will not used it

API_AustralianOpen.drop(columns=['fields.location'], inplace=True)
```

```python
#We create a new dataframe for the 'fields.winner'. We want only the 'Last Name' so when comparing with WEB data, we have the same data.

new1 = API_AustralianOpen['fields.winner'].str.split(" ", n=1, expand = True)

new1
```

```python
#We create the 'Winner' column with index[0] of previous step. It will be the last name of the player/winner

API_AustralianOpen['Winner']=new1[0]
```

```python
#We create the 'Year' column to only have the Year as YYYY

API_AustralianOpen['Year']= pd.DatetimeIndex(API_AustralianOpen['fields.date']).year
```

```python
API_AustralianOpen
```

```python
#We drop the column 'fields.winner' and 'fields.date' as we will not used it

API_AustralianOpen.drop(columns=['fields.winner', 'fields.date'], inplace=True)
```

```python
#We sort the values by 'Year' and we store it

API_AustralianOpen.sort_values(by='Year', ascending=False, inplace=True)
```

```python
#We print the results for checking everything is fine

API_AustralianOpen
```

```python
#We export the data as CSV file as requested

API_AustralianOpen.to_csv(r'C:\Users\victo\Desktop\IronHack\PR03-project-web\output\API_AustralianOpen.csv', index=False)
```

# WEB SCRAPING

```python
#Setting the url from the WEB selected

URL = 'https://en.wikipedia.org/wiki/List_of_Grand_Slam_men%27s_singles_champions'

page = requests.get(URL)
```

```python
#Checking the status code return is correct

page.status_code
```

```python
#We parse the html by using BeautifulSoup library

soup = BeautifulSoup(page.content, "html.parser")

soup
```

```python
#We identify the table as previously inspecting the data. Table is under class 'wikitable sortable'

table = soup.find('table', class_="wikitable sortable")

table
```

```python
#We convert the table into a dataframe by using 'pd.read_html'

table1 = pd.read_html(str(table))
table1
```

```python
#We convert the previous 'dataframe' into a new one by setting index=[0]
WEB_FrenchOpen = table1[0]
WEB_FrenchOpen
```

```python
#We leave the dataframe with the columns only for our analysis

WEB_FrenchOpen = WEB_FrenchOpen[['Year', 'French Open']]
```

```python
#We filter the results from 2001 to 2016 into two steps

WEB_FrenchOpen = WEB_FrenchOpen[WEB_FrenchOpen['Year']>=2001]
WEB_FrenchOpen = WEB_FrenchOpen[WEB_FrenchOpen['Year']<=2016]
```

```python
#We print the results for checking if the steps applied are correct

WEB_FrenchOpen
```

```python
#We replace what is '( )' for empty

WEB_FrenchOpen['French Open'] = WEB_FrenchOpen['French Open'].str.replace(r'\W\d+\W+\d+\W','')
```

```python
#We print the results for checking that it has been replaced

WEB_FrenchOpen
```

```python
#We create a new dataframe for the 'Australian Open'. We want only the 'Last Name' so when comparing with API data, we have the same data.

new = WEB_FrenchOpen['French Open'].str.split(" ", n=1, expand = True)
```

```python
#We create the 'Winner' column with index[0] of previous step. It will be the last name of the player/winner

WEB_FrenchOpen['Winner']=new[1]
```

```python
#We drop the column 'Australian Open' as we will not used it

WEB_FrenchOpen.drop(columns=['French Open'], inplace=True)
```

```python
#We sort by 'Year', ascending=False

WEB_FrenchOpen.sort_values(by='Year', ascending=False, inplace=True)
```
