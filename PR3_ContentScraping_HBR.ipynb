```python
from bs4 import BeautifulSoup
import requests
import json
import pandas as pd
from pandas import json_normalize
```

### Preparar el código:

```python
url = "https://hbr.org/search?search_type=search-all&term=hr+analytics&loaded=10&sort=publication_date"
res_hbr = requests.get(url)
```

```python
html = res_hbr.text
html
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
soup
```

```python
#Vas a buscar una muestra del texto sobre la que después podrás iterar:
results = soup.find_all("stream-item", attrs = {"class":"stream-item overflow-hidden"})
```

```python
#Tenemos 112 entradas
len(results)
```

### Ahora que ya tenemos todos los elementos que queremos para una ficha, vamos a aplicarlo a todas las entradas:

```python
entries = []
for result in results:
    titulo = result["data-title"]
    topic = result["data-topic"]
    tipo = result["data-content-type"] 
    desc = result["data-summary"]
    url = "https://hbr.org" + result.find("a")["href"]
    price = results[0]["data-list-price"]
    entries.append([titulo, topic, tipo, desc, url, price])
```

```python
entries
```

### Monta el dataframe:

```python
import pandas as pd
```

```python
df = pd.DataFrame(entries, columns=["Título", "Topic", "Tipo", "Descripción", "url", "Precio"])
```
