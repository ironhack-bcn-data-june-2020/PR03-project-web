## I used the https://footystats.org/api/ to get all the data

```python
url = 'https://api.footystats.org/league-matches?key=example&league_id=2012'
#response = requests.get(api_url)
fp = urllib.request.urlopen(url)
mybytes = fp.read()
response = mybytes.decode("utf8")
fp.close()
rates = json.loads(response)
```

## I ordered the Data Frame by Wins and I created two new columns. the number of games played and the probability of winning and Tie per team

```python
df.sort_values("Wins", ascending= False, inplace= True)
df["Played"]= df["Wins"]+ df["Lost"] + df["Tie"]
df["Proba_of_winning"] = (df["Wins"] / df["Played"])*100
df["Proba_of_Tie"]=(df["Tie"] / df["Played"])*100
df
```
