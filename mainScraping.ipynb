## The tools that we are going to use in this project

```python
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup as bs
```

## The website URL that we are going to work with

```python
url = 'http://sofifa.com/players?offset=0'
```

## I created a function that extract the information from the website

```python
def sofifa(url):
    r = requests.get(url)
    response = r.content
    soup = bs(response, 'lxml')
    return soup
```

```python
sofifa= sofifa(url)
```

```python
sofifa.text
```

## I extracted the body of the whole website code

```python
results = sofifa.find_all('tbody', {'class': 'list'})
results
```

## I extracted from the website the name of the players

```python
players = sofifa.find_all('a', {'class': 'tooltip'})
players
```

```python
first_player= players[0]
first_player
```

## I created a list that contain the names of the players

```python
players_name=[]
for x in players:
    players_name.append(x.text)
players_name = players_name[:-4]
players_name
```

## I extracted from the website the team and season, where the per players are at the moment

```python
team=sofifa.find_all('td',{"class":"col-name"})
team[1].find("a").text
```

## I created a list that contain the team and season of the players at the moment

```python
team_name=[]
for x in team:
    
    team_name.append(x.text.replace("\n",""))
team_name=team_name[1::2]
team_name
```

## I extracted the age per player from the website

```python
edad = sofifa.find_all('td',{"class":"col-ae"})
edad[0].text
```

## I created a list that contain the age per player

```python
players_age=[]
for x in edad:
    players_age.append(x.text)
players_age
```

## I extracted the value per player from the website

```python
value = sofifa.find_all('td',{"class":"col-vl"})
value[0]
```

## I created a list that contain the value per player

```python
players_value=[]
for x in value:
    players_value.append(x.text.replace("M"," Mill"))
players_value
```

## I extracted all the salaries from the website

```python
salary = sofifa.find_all('td',{"class":"col-wg"})
salary[0]
```

## I created a list that contain all the salaries per player

```python
players_salary=[]
for x in salary:
    players_salary.append(x.text.replace("K"," Mil"))
players_salary
```

## I created a dictionary with the all variables

```python
table={"Players_Name":players_name,"Players_Age":players_age,"Team_Name&Season":team_name,"Players_Value":players_value,"Player_week_Salary":players_salary}
```

## I created a data frame to show the final results of the value per player in the market

```python
df_table=pd.DataFrame(table)
df_table
```

## In this code if you change the variable total_n2 will show you from which position of the table you wants to see the players, for example, if you tells from 1 will shows you from the first position of the table and if you tell 30 it will start to show you the players from the 30 position forward, It will take longer than usual if you put a number over 10.

```python
total_n2=1 #if I put a number over a 10 the programs takes long to load all the players from the website 
for page in range(total_n2):
    url=f'http://sofifa.com/players?offset={page}'
    r= requests.get(url)
    response = r.content
    sofifa = bs(response, 'lxml')
    players = sofifa.find_all('a', {'class': 'tooltip'})
    players_name=[]
    for x in players:
        players_name.append(x.text)
    players_name = players_name[:-4]
    team=sofifa.find_all('td',{"class":"col-name"})
    team_name=[]
    for x in team:
        team_name.append(x.text.replace("\n",""))
    team_name=team_name[1::2]
    edad = sofifa.find_all('td',{"class":"col-ae"})
    players_age=[]
    for x in edad:
        players_age.append(x.text)
    value = sofifa.find_all('td',{"class":"col-vl"})
    players_value=[]
    for x in value:
        players_value.append(x.text.replace("M","").replace("€","").replace("275K","0.275"))
    
    salary = sofifa.find_all('td',{"class":"col-wg"})
    players_salary=[]
    for x in salary:
        players_salary.append(x.text.replace("K","").replace("€",""))
    table={"Players_Name":players_name,"Players_Age":players_age,"Team_Name&Season":team_name,"Players_Value_M":players_value,"Player_week_Salary_K":players_salary}
    df_table1=pd.DataFrame(table)
df_table1
#Players_Value_M valor de los jugadores en millones(M)
#Player_week_Salary_K salario de los jugadores en miles por semana
```

## I saved all the information in a CSV file

```python
df_table1.to_csv('player.csv', index=False, encoding='utf-8')
```
