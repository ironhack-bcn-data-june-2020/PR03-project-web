```python
import requests
```

```python
url = 'https://www.oscars.org/oscars/ceremonies'
```

```python
awards = requests.get(url)
urltext = awards.text
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(urltext, 'html.parser')
years = soup.findAll('div', {"class":"years"})
```
