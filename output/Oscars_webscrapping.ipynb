# THE OSCARS

```python
import requests
```

```python
url = 'https://www.oscars.org/oscars/ceremonies/1940'
```

```python
awards = requests.get(url)
urltext = awards.text
```

```python
print(urltext)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(urltext, 'html.parser')
decades = soup.findAll('div', {"class":"decades"})
```
