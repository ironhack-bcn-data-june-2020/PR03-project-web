# Wikiscrapping

```python
import requests
from bs4 import BeautifulSoup
```

### Getting Billy Wilder info

```python
url = requests.get('https://en.wikipedia.org/wiki/Billy_Wilder').text
```

```python
soup = BeautifulSoup(url,'html.parser')
print(soup.prettify())
```

### Awards table

```python
# Looking for the awards list, it happens to be placed in an irregular table
awards_table = soup.find('table',{'class':'wikitable plainrowheaders sortable'})
awards_table
```

```python
"""Trying to extract the information of awards, unfortunately, it was impossible for me to extract the "result" column
with two diferent classes "td class= yes table-yes2" for "won" and "td class=no table-no2" for "nominated"
```

```python
# Making a single list for all the rows"
rows = awards_table.find_all('a')
rows = [row.text.strip() for row in rows]
rows
```

```python
"""Since I got an arranged list by years of awards, I tried to delete the ones which are no Academy Awards (Oscars)"""

only_oscars = rows[:77]
print(only_oscars)
```

```python
#Following some Paula advice, trying to split a list into sublists of given length

from itertools import accumulate 

clean_list = []

# list of length in which we have to split by year, movie and nominations
length_to_split = [2, 5, 8, 6, 6, 5, 6, 5, 4, 6, 4, 6, 7, 4, 3]
  
# Using islice 
output= [only_oscars[x - y: x] for x, y in zip( 
          accumulate(length_to_split), length_to_split)] 

clean_list.append(output)
print(clean_list)
```

```python
for inner_clean_list in clean_list:
    for item in inner_clean_list:
        print(item)
```
