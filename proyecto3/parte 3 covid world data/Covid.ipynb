## 1- Importación de librerias y llamada a la url


```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
headers = {
        'USER_AGENT': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64; rv:47.0) Gecko/20100101 Firefox/47.0'
    }
```

```python
url = "https://www.worldometers.info/coronavirus/"
```

```python
# Hago uns petición Get a la url 
request = requests.get(url, headers=headers)
```

```python
# Creo un soup de BeautifuSoup con el contenido de la página
# Parse de html
soup = BeautifulSoup(request.content, "html.parser")
```

## Extración de datos con Beautiful Soup

```python
# Escojo el id de la tabla que quiero scrapear
# Estraigo el código html de esa tabla únicamente
table_covid = soup.find("table", attrs={"id": "main_table_countries_today"})
```

```python
# Creo una variable general buscando los elementos tr de la tabla
head = table_covid.find_all("tr")
```

```python
head
```

```python
# Reemplazo los saltos de linea y le aplico un strip() a todos los th de la variable general head
# Todos esos datos los guardo en una lista usando un list comprehension
# Estos datos serán los nombres de las columnas del futuro Data Frame
lst_head = [th.text.replace("\n", "").strip() for th in head[0].find_all("th")]
```

```python
# Compruebo que se ha cargado correctamente
lst_head
```

```python
# Creo una variable general body que contiene los datos de lo que serán las filas 
# Buscando todos los tr de la tabla html que los contiene en la página web
body_covid = table_covid.tbody.find_all("tr")
```

```python
# Creo la variable rows a la que mediante un bucle for le pido que añada todos los datos de body enumerados
rows = []
for r in range(1, len(body_covid)):
    row = [tr.text.replace("\n", "").strip()for tr in body_covid[r].find_all("td")]
    rows.append(row)
```

```python
# Crompobamos que se ha cargado bien
rows
```

```python
# Creo el Data Frame con los datos scrapeados
df_data_covid = pd.DataFrame(rows, columns=lst_head)
# Crompuebo que se ha cargado correctamente con head
df_data_covid.head(10)
```
